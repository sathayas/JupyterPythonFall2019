# 3. Axes, titles, etc
<hr style="height:1px;border:none" />

Notice that your plot does not have a title, and axes do not have any label. You can add the title by **`plt.title()`**. You can also add x- and y-axis labels by **`plt.xlabel()`** and **`plt.ylabel()`**, respectively. For example,

[`<PlotTitleLabels.py>`](https://github.com/sathayas/PythonClassFall2019/blob/master/plotExamples/PlotTitleLabels.py)

````python
import matplotlib.pyplot as plt
import numpy as np

# Generating simple data
t = np.linspace(0,4*np.math.pi,64)
y = np.sin(t)

# plot with title and axis labels
plt.plot(t, y, 'b-')
plt.title('Sine function')
plt.ylabel('sin(t)')
plt.xlabel('Angle t (radian)')
plt.show()
````

````python
plt.plot(t, y, 'b-')
plt.title('Sine function')
plt.ylabel('sin(t)')
plt.xlabel('Angle t (radian)')
plt.axis([0,13,-2,2])
plt.show()
````

This re-adjusts the x-axis from 0 to 13, and the y-axis from -2 to 2.

# 4. Legend
<hr style="height:1px;border:none" />

If you have multiple series on your plot, then you may want to add a legend, explaining which series corresponds to which. To do so, you need to add a label to each series you plot. Then the **`plt.legend()`** function automatically gather the label and plot symbol/color information and generates a legend. For example,

[`<PlotLegend.py>`](https://github.com/sathayas/PythonClassFall2019/blob/master/plotExamples/PlotLegend.py)

````python
import matplotlib.pyplot as plt
import numpy as np

# Generating simple data
t = np.linspace(0,4*np.math.pi,64)
y = np.sin(t)

# plotting the data with labels
plt.plot(t, y, 'b-', label='Blue curve')
plt.plot(t, y+0.2, 'g.', label='Green dots')
plt.plot(t, y+0.4, 'ro', label='Red circles')
# and creating a legend
plt.legend()
plt.show()
````

Here, each **`plt.plot()`** function has an input parameter **`label`**, a string describing what this series is about. 

The location of the legend is on the top right by default. You can change this by specifying the parameter **`loc`** in the `plt.legend()` function. For example,

````python
# plotting the data with labels
plt.plot(t, y, 'b-', label='Blue curve')
plt.plot(t, y+0.2, 'g.', label='Green dots')
plt.plot(t, y+0.4, 'ro', label='Red circles')
# and creating a legend
plt.legend(loc=10)
plt.show()
````

Will result in the legend located on the right side of the plot. The input parameter `loc` takes an integer indicating a location. Here is a list of `loc` and the corresponding locations.

# 5. Saving a figure
<hr style="height:1px;border:none" />

So far, we learned how to generate a simple scatter plot. Now you want to save the figure so that you can use it later for your paper, presentation, or grant proposal. To do so, you can use the **`plt.savefig()`** function. Here is an example.

[`<SaveFigure.py>`](https://github.com/sathayas/PythonClassFall2019/blob/master/plotExamples/SaveFigure.py)

````python
import matplotlib.pyplot as plt
import numpy as np

# Generating simple data
t = np.linspace(0,4*np.math.pi,64)
y = np.sin(t)

# The range of the axes are controlled now
plt.plot(t, y, 'b-')
plt.title('Sine function')
plt.ylabel('sin(t)')
plt.xlabel('Angle t (radian)')
plt.axis([0,13,-2,2])
plt.savefig('SinPlot.png', dpi=300)
plt.show()
````

In this case, the figure is saved as **`SinPlot.png`**, located in the same directory as this program. By default, **`plt.savefig()`** saves the figure in the **`.png`** format. Once you have your figure as an image file, you can use it for your papers or presentations. 

If you ever need to generate a figure with high resolution (for publication, for example), then you can adjust the **`dpi`** parameter in the `plt.savefig()` function. For example,
```python
plt.savefig('SinPlot.png', dpi=600)
```
Produces a figure with 600 dpi.


# 6. Multiple subplots
<hr style="height:1px;border:none" />

You can save individual panels for a figure with multiple panels. Or, you can create a figure with multiple sub-plots. Here is an example.

[`<SubplotExample.py>`](https://github.com/sathayas/PythonClassFall2019/blob/master/plotExamples/SubplotExample.py)

````python
import matplotlib.pyplot as plt
import numpy as np

# Generating simple data
t = np.linspace(0,4*np.math.pi,64)
y = np.sin(t)
z = np.cos(t)
w = np.tan(t)

# first, starting a figure
plt.figure(figsize=[6,6])

# first subplot
plt.subplot(221)
plt.plot(t, y, 'b-')
plt.title('Sine function')
plt.ylabel('sin(t)')
plt.xlabel('Angle t (radian)')
plt.axis([0,13,-1.1,1.1])

# second subplot
plt.subplot(222)
plt.plot(t, z, 'b-')
plt.title('Cosine function')
plt.ylabel('cos(t)')
plt.xlabel('Angle t (radian)')
plt.axis([0,13,-1.1,1.1])

# third subplot
plt.subplot(223)
plt.plot(t, w, 'b-')
plt.title('Tangent function')
plt.ylabel('tan(t)')
plt.xlabel('Angle t (radian)')
plt.axis([0,13,-1.1,1.1])

plt.show()
````

In this program, before starting to plot anything, you run 
```python
plt.figure(figsize=[6,6])
```
This is to create a figure of size 6in x 6in. Before each subplot, you see
```python
plt.subplot(221)
```
This means you will be creating a subplot, and whatever you will execute from this point on will be part of the subplot **`221`**. What does **`221`** mean? The first two numbers (2 and 2, in this example) means there will be a plot matrix of 2 rows and 2 columns. The third number means the plot number 1. In a plot matrix of 2x2, there are 4 subplots. The subplot 1 is located at the 1st row, 1st column of this matrix. The other subplots, 2, 3, and 4 in this case, fills the matrix row by row. In this particular example, there are 4 places in the plot matrix, but you are using only 3 of them. 

When you examine the resulting plot matrix, you notice that labels and titles are overlapping between subplots. This is embarrassing! This occurs because there is not enough space between subplots. To adjust this, you can include this in your program.
```python
plt.subplots_adjust(wspace=0.4, hspace=0.4)
```
This function adjusts spaces between subplots, horizontally (**`wspace`**) and vertically (**`hspace`**). The values for these parameters are determined by trial and error. In this case, 0.4 for both seem to work very well, as you can see the adjusted figure.

````python
# first, starting a figure
plt.figure(figsize=[6,6])

# first subplot
plt.subplot(221)
plt.plot(t, y, 'b-')
plt.title('Sine function')
plt.ylabel('sin(t)')
plt.xlabel('Angle t (radian)')
plt.axis([0,13,-1.1,1.1])

# second subplot
plt.subplot(222)
plt.plot(t, z, 'b-')
plt.title('Cosine function')
plt.ylabel('cos(t)')
plt.xlabel('Angle t (radian)')
plt.axis([0,13,-1.1,1.1])

# third subplot
plt.subplot(223)
plt.plot(t, w, 'b-')
plt.title('Tangent function')
plt.ylabel('tan(t)')
plt.xlabel('Angle t (radian)')
plt.axis([0,13,-1.1,1.1])

# adjusting the space between subplots
plt.subplots_adjust(wspace=0.4, hspace=0.4)

plt.show()
````

# 7. Bar graphs
<hr style="height:1px;border:none" />

You can plot bar graphs with `matplotlib` too. Here is a simple example.

[`<BarGraphExample.py>`](https://github.com/sathayas/PythonClassFall2019/blob/master/plotExamples/BarGraphExample.py)

````python
import numpy as np
import matplotlib.pyplot as plt

meanData = np.array([20, 35, 30, 35, 27])
stdData = np.array([2, 3, 4, 1, 2])
x = np.arange(1,6)

# just plotting bars
plt.bar(x, meanData)
plt.show()
````

Notice that you need to specify the location of the bar on the x-axis as the first parameter. The second parameter describe the height of the bar. 

You can also modify the plot with different parameters. You can modify the width of bars by providing the third parameter after the height, and you can modify the color of bars with the **`color`** parameter. 

````python
# bar width and colors
plt.bar(x, meanData, 0.4, color='m')
plt.show()
````

You can also add error bars using the **`yerr`** parameter.

````python
# error bars added
plt.bar(x, meanData, 0.4, yerr=stdData, color='m')
plt.show()
````

Now, we can format axes and labels so that the plot looks nice. 

````python
# formatting the x- and y-axis, adding the title
plt.bar(x, meanData, 0.4, yerr=stdData, color='m')
plt.xlabel('Groups')
plt.ylabel('Scores')
plt.title('Scores by groups')
plt.xticks(x, ['Patients', 'Controls', 'Parents', 'Siblings', 'Placebo'])
plt.show()
````

Here, the group labels on the x-axis is generated by the **`plt.xticks()`** function. You provide a series of x-coordinates that are centers of the bars. You can specify names of the groups by a list of strings.

## Multi-group bar graphs

Say, you want to plot bar graphs of two groups by side-by-side for a comparison. You can do that by plotting two sets of bars.

[`<MultiBarGraphExample.py>`](https://github.com/sathayas/PythonClassFall2019/blob/master/plotExamples/MultiBarGraphExample.py)

````python
import numpy as np
import matplotlib.pyplot as plt

# data to be plotted
means_men = (20, 35, 30, 35, 27)
std_men = (2, 3, 4, 1, 2)
means_women = (25, 32, 34, 20, 25)
std_women = (3, 5, 2, 3, 3)

# paramters
bar_width = 0.3
x = np.arange(len(means_men))

# plotting bars
plt.bar(x, means_men, bar_width, yerr=std_men,
        color='b', label='Men')
plt.bar(x+bar_width, means_women, bar_width, yerr=std_women,
        color='r', label='Women')

# formatting and labeling the axes and title
plt.xlabel('Groups')
plt.ylabel('Scores')
plt.title('Scores by groups')

# formatting ticks for the x-axis
plt.xticks(x+0.5*bar_width, ['Patients', 'Controls', 'Parents', 'Siblings', 'Placebo'])

# adding the legend
plt.legend()

# and we are done!
plt.show()
````

Here, you need to be careful about the x-coordinates of the second set of bars. They need to be shifted to the right by a bar width, so that there is no overlap between the first and second groups of bars.
